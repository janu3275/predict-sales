```python
data = df.groupby(['shop_id','item_id']).item_cnt_day.count().reset_index()
```

```python
data['monthly'] = (data['item_cnt_day']/10)
```

```python
data.rename(columns = {'item_cnt_day':'ten_months'},inplace= True)
print(data.head())
```

```python
data2 = pd.merge(data,items,left_on='item_id',right_on = 'item_id')
data2 = data2.drop('item_name',axis = 1)
print(data2.head())
```

```python
train_data = data2.drop(['ten_months','monthly'],axis =1)
print(train_data.head())
```

```python
train_labels = data['monthly']*10
```

```python
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier()
forest.fit(train_data2,train_labels)
print(forest.score(train_data2,train_labels))
```
